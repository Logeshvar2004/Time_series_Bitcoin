```python
bitcoin_df['Date']=pd.to_datetime(bitcoin_df['Date'])
```

```python
bitcoin_df.corr()
```

```python
df=bitcoin_df.corr()
df['Close'].sort_values(ascending=False)
```

```python
from autots import AutoTS
model= AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
```

```python
model=model.fit(bitcoin_df,date_col='Date',value_col='Close')
```

```python
import pickle
with open(r'D:\Coding\python\Bit\autots_model.pkl','wb') as model_file:
    pickle.dump(model, model_file)
```

```python
pred = model.predict()
```

```python
forecast= pred.forecast
```

```python
print(forecast)
```

```python
forecast
```

```python
plt.plot(forecast.index, forecast.values, marker='o', ls='-')
plt.title('Bitcoin Price Over Time')
plt.xlabel('month')
plt.xticks(rotation=90)
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```
